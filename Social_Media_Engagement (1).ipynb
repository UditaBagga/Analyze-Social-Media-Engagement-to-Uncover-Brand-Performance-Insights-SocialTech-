HYPOTHESIS TESTING

HYPOTHESIS 1

H0 : Content type does NOT affect engagement.

H1 : Content type affects engagement.

```python
df['engagement_level'] = df['engagement_rate'] > df['engagement_rate'].median()
table = pd.crosstab(df['content_type'], df['engagement_level'])
print(table)
```

```python
from scipy.stats import chi2_contingency
chi2, p, dof, expected = chi2_contingency(table)
print("p-value:", p)
```

INTERPRETATION : Content type significantly influences engagement. Different content formats such as image, video, text, and mixed posts generate different levels of audience interaction.

HYPOTHESIS 2

H0: Platform and engagement level are independent

H1: Platform influences engagement level

```python
from scipy.stats import chi2_contingency

df["high_engagement"] = (
df["engagement_rate"] >
df["engagement_rate"].median()
).astype(int)

table = pd.crosstab(df["platform"], df["high_engagement"])

chi2,p,dof,expected = chi2_contingency(table)

print(p)
```

INTERPRETATION : Platform significantly influences engagement levels. Audience interaction varies across different social media platforms.

HYPOTHESIS 3

H0: Disclosure type does not affect engagement.

H1: Disclosure type affects engagement.

```python
df['engagement_level'] = df['engagement_rate'] > df['engagement_rate'].median()

table = pd.crosstab(df['disclosure_type'], df['engagement_level'])
print(table)

from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(table)
print("p-value:", p)
```

INTERPRETATION : Disclosure type does not have a statistically significant effect on engagement. Explicit or implicit disclosures generate similar engagement levels.

HYPOTHESIS 4

H0: Posting day has no effect on engagement

H1: Engagement varies across posting days

```python
# 1. Create engagement category (High / Low)
df['engagement_level'] = df['engagement_rate'] > df['engagement_rate'].median()

# 2. Create contingency table
table = pd.crosstab(df['post_day'], df['engagement_level'])
print(table)

# 3. Chi-Square test
from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(table)
print("p-value:", p)
```

INTERPRETATION : Posting day significantly affects engagement. Audience interaction varies depending on the day content is posted.

HYPOTHESIS 5

H0: Engagement is independent of audience gender

H1: Engagement differs across gender groups

```python
table = pd.crosstab(
    df["audience_gender_distribution"],
    df["high_engagement"])

from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(table)

print("Chi-square:", chi2)
print("P-value:", p)
```

INTERPRETATION : Audience gender distribution does not significantly influence engagement levels.

HYPOTHESIS 6

H0: Follower count does not impact engagement

H1: Higher followers increase engagement

```python
sample_size = min(len(high_follow), len(low_follow), 30)

high_sample = high_follow.sample(sample_size, random_state=42)
low_sample = low_follow.sample(sample_size, random_state=42)
```

```python
from scipy.stats import ttest_ind
t_stat, p_value = ttest_ind(
    high_sample,
    low_sample,
    equal_var=False)
print("T-statistic:", t_stat)
print("P-value:", p_value)
```

INTERPRETATION : Follower count does not show a statistically significant impact on engagement in the sampled data.

K MEAN CLUSTERING

```python
features = df[['engagement_rate','follower_count','total_engagement']]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(features)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = []

for i in range(1,10):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,10), wcss, marker='o')
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.title("Elbow Method")
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)

df['cluster'] = kmeans.fit_predict(X_scaled)
```

```python
cluster_order = df.groupby('cluster')['engagement_rate'].mean().sort_values().index

mapping = {cluster_order[i]: i for i in range(len(cluster_order))}

df['cluster'] = df['cluster'].map(mapping)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

sns.scatterplot(
    x='follower_count',
    y='engagement_rate',
    hue='cluster',
    palette=['blue','orange','green'],
    data=df
)

plt.title("Post Engagement Clusters")

plt.show()
```

Cluster 0 — Low Engagement Posts

Cluster 1 — Medium Engagement Posts

Cluster 2 — High Engagement Posts

The clustering analysis reveals that engagement rate does not increase proportionally with follower count. Some smaller creators demonstrate stronger engagement compared to larger influencers, highlighting the importance of micro-influencers in social media marketing strategies.
